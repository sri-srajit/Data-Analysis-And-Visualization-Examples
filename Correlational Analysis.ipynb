# Correlations in Python
* It ranges between -1 to 1


## Import all packages

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
%matplotlib inline
```

```python
#set the numbers to repeat every time
np.random.seed(1)
```

```python
#100 random numbers between 0 and 50
x = np.random.randint(0,50,100)
```

```python
#positive correlation with some noise
#normal is used to distribute normal values randomly
y = x + np.random.normal(0,5,100)
```

```python
#show the correlation
#positive correlation - y increases with increase in x
#negative correlation - y decreases with increase in x
#It is show in form of a 2x2 matrix
#[[xx,xy],[yx,yy]]
```

```python
#corrcoef() is used to find correlation
np.corrcoef(x,y)
```

* A strong positive correlation

```python
#plot the graph
plt.scatter(x,y)
plt.show()
```

## Negative Correlation

```python
x = np.linspace(0,100,10)
```

```python
y = -5*x + 3
```

```python
np.corrcoef(x,y)
```

```python
plt.scatter(x,y,color = 'red',linewidths = 0.1)
plt.show()
```

## Correlation Matrix

```python
df = pd.DataFrame({'a':np.random.randint(0,20,1000)})
```

```python
# positively correlated with 'a'
df['b'] = df['a'] + np.random.normal(0,20,1000)
```

```python
#negatively correlated with 'a'
df['c'] = 100 - df['a'] + np.random.normal(0,5,1000)
```

```python
#not correlated with 'a'
df['d'] = np.random.randint(0,20,1000)
```

```python
#show the correlation
df.corr()
```

```python
#plot the graph to visualize
#scatter matrix is in pandas.plotting so import it
from pandas.plotting import scatter_matrix
```

```python
scatter_matrix(df)
plt.show()
```

```python
sb.pairplot(df)
plt.show()
```
