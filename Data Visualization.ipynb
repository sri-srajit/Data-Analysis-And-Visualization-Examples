## Q.2 
- Plot a Sine Wave with color code with "black" color and style as dashdot, also mark "x" and "y" label and give a suitable title to it

```python
x = np.linspace(-10,10,100)
y = np.sin(x)
plt.plot(x,y,color = 'black',linestyle='dashed')
plt.xlabel('X axis',size = 20)
plt.ylabel('Y axis',size = 20)
plt.title('Sine wave Curve',size = 30)
plt.show()
```

## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
#create the data
data = {'Store':['Delhi','Mumbai','Chennai','Kolkata','Hyederabad'],'Clothing':[82,78,48,46,39],'Equipment':[48,31,51,29,18],'Accessories':[15,20,14,13,9]}
data
```

```python
#store all values in numpy array 
cities = data['Store']
cloth = np.array(data['Clothing'])
equip = np.array(data['Equipment'])
accs = np.array(data['Accessories'])
var = np.array([x for x,_ in enumerate(cities)])
var = var*2
```

```python
plt.bar(var,cloth,width = 0.5,label = 'Clothing',color = 'c')
plt.bar(var+0.5,equip,width = 0.5,label = 'Equipment',color = 'm')
plt.bar(var+1.0,accs,width = 0.5,label = 'Accessories',color = 'violet')
plt.xlabel('City Names',size = 15)
plt.ylabel('Data',size = 15)
plt.xticks(var+0.5,cities,rotation = 90)
plt.legend(loc = 'upper right')
plt.title('Store Data',size = 20)
plt.show()
```

## Q.5 For Titanic Dataset
- Load Dataset using Seaborn
- Plot a Bar Graph of Survival of Passenger against the Passenger Class.
- Use proper "hue" attribute, and palette color. Mark 'x' and 'y' label with proper title.
- Use legend located at upper left corner

```python
#look for the dataset
sb.get_dataset_names()
```

```python
#load the dataset into memory
titanic = sb.load_dataset('titanic')
```

```python
#Know the type of dataset
titanic.head()
```

```python
titanic['alive'].replace(['yes','no'],[1,0],inplace = True)
```

```python
titanic.head()
```

```python
g = sb.barplot(x = 'pclass',y = 'alive',data = titanic,hue = 'pclass',palette = 'bright')
g.set(xlabel = 'Passenge Class',ylabel = 'Life Chance')
g.legend(loc = 'upper left')
```
